```python
# Importing packages 

import pandas as pd
import numpy as np
```

```python
# Reading data in from Worlds File

raw_world_data = pd.read_csv('./2017-worlds-match-data-OraclesElixir-2017-11-05.csv')
print(raw_world_data.shape)
raw_world_data.head()
```

```python
# Checking out our data 
raw_world_data.describe()
```

```python
# Since count is the same for all lets check to make sure there are no 
# out of bounds values

temp_raw_null = raw_world_data.isnull().any()
temp_raw_null[temp_raw_null == True]
```

```python
# There was a null value in Visible ward clear rate. Lets see what that looks like
raw_world_data['visiblewardclearrate']
```

```python
# This appears to be team values that are not player related so 
# All team rows have Team as position so lets select those
team_filter = raw_world_data['position'] == 'Team'
team_filter.head(20)
```

```python
# Putting only team data in the team_data df
team_data = raw_world_data[team_filter]
team_data.head()
```

```python
# Lets do the same but this time but with the player data 
player_filter = raw_world_data['position'] != 'Team'
player_data = raw_world_data[player_filter]
player_data.head()
```

```python
# Lets delete some useless data
player_data = player_data.drop(player_data.columns[1:9], axis = 1)
```

```python
# Lets check agian to see if any null values
temp_null = player_data.isnull().any()
temp_null[temp_null == True]
```

```python
# Fill null values with -1 for not taken
player_data = player_data.fillna(value = -1)
```

```python
# Final check to ensure no null values
null_data = player_data.isnull().any()
null_data[null_data == True]
```

```python
# No true therefore all values are filled
# Lets take another look at the player data

player_data.describe()
```

```python
# Lets compute the pairwise corr for the columns
player_corr = player_data.corr()

# The data here is messy so lets only pull out those values that 
# Are strongly correlated I.E. >=0.5 or <= -0.5
# Since we are selecting only those highly corr we need to create a corr df
pos_corr_filter = player_corr >= 0.5 
neg_corr_filter = player_corr <= -0.5
pos_player_corr = player_corr[pos_corr_filter]
neg_player_corr = player_corr[neg_corr_filter]
```

```python
# Lets see what the data looks like now count gives a good aprox for
# How often a stat is correlated with other stats.  
pos_player_described = pos_player_corr.describe()
max(pos_player_described.loc['count'])
```

```python
# The max count is 14.0 so lets see which col that is
temp_count = pos_player_described.loc['count'] == 14.0
temp_count[temp_count == True]
```

```python
# earnedgpm removes the starting gold and gold gen per min 
# This stat seems to mean that teams that earn more gold win more, as
# to be expected lets look at neg now
neg_player_described = neg_player_corr.describe()
max(neg_player_described.loc['count'])
```

```python
# What does 11 relate to
temp_count = neg_player_described.loc['count'] == 11.0
temp_count[temp_count == True]
```

```python
# Well that is strange wards per minute and share of team wards placed
# seem to negativly correlate more often lets double check that
neg_player_described[['wpm','wardshare']]
```

```python
# both have a strong negative mean between -0.5 and -0.8 lets look at wins
# in comparison to wpm and wardshare
player_data['wpm'].describe()
```

```python
# Here we can see that the 50% value is 0.57 and the mean is 0.69
# Lets see if players that placed <= 0.44 lost more games than won

#Selecting those plyers whose wpm <= 0.44
low_wpm = player_data['wpm'] <= 0.44
low_player_wpm = player_data[low_wpm]

#Check to make sure the values are less than 0.44 at max
max(low_player_wpm['wpm'])
```

```python
# Lets get a count of the result now 1 = win so we can just count the wins
# and then compare to the avg

# First we have to count the num of wins from the entire tournament
total_wins = player_data['result'].count()

#Next lets look at the wins from our low wpm players
low_wins = low_player_wpm['result'].count()

print('Total wins at worlds:',total_wins,'\n','Low player wins:',low_wins)
```

```python
# First lets see who these low warding players are (SHAME ON THEM)
bad_warders = low_player_wpm.groupby(['player','position']).size().reset_index(name='counts')
bad_warders.sort_values(by='counts',ascending=False)

```

```python
# That gives us an idea that they won less games but that doesn't show
# If there is a correlation lets see
low_wpm_filter = low_player_wpm.corr()['wpm'] 
low_wpm_filter['result']
```
