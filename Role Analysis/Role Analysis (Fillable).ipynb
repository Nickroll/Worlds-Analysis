```python
# Here we are going to look at the Role position from worlds
# This will follow the same template as my mid analysis but only looking
# at gdat10 and gdat15 
import pandas as pd
import numpy as np
import plotly.plotly as py
import plotly.graph_objs as go
import plotly.offline as offline

offline.init_notebook_mode(connected = True)
```

```python
player_data = pd.read_csv('../Data/Player_specific_data.csv')
player_data.head()
```

```python
#Getting only Role players
Role_mask = player_data['position'] == 'Role'
Roles = player_data[Role_mask]
Roles.head()
```

```python
# Lets group people together
Role_stats = Roles.groupby(by ='player').describe()
Role_stats
```

```python
# Have to get the avg stats only 
Role_stats_avg = Role_stats.iloc[:,Role_stats.columns.get_level_values(1) == 'mean']
Role_stats_avg.columns = Role_stats_avg.columns.droplevel(1)
Role_stats_avg.head()
```

```python
# Lets graph the gdat10 and gdat15 
temp_sort = Role_stats_avg.sort_values(by = 'Var1')
var1 = list(temp_sort['Var1'].values)
var2 = list(temp_sort['Var2'].values)
players = list(temp_sort.index)

trace1 = go.Bar(x = players, y = var1, name = 'Var1')
trace2 = go.Bar(x = players, y = var2, name = 'Var2')

data = [trace1,trace2]
layout = go.Layout(barmode = 'grouped',
                  title = 'Average Role Var1 vs. Var2',
                  yaxis = dict(title = 'Var1'),
                  xaxis = dict(tickangle = 90)
                  )

fig = go.Figure(data = data, layout = layout)

#offline.plot(fig, filename ='Final_Var1_vs_Var2_Diff.html')

```
